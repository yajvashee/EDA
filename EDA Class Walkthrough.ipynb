```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load the Dataset

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(columns =['Unnamed: 0'],inplace = True)
```

```python
df.drop(columns =['PassengerId','Name'],inplace = True)
```

## Univariate Analysis

```python
df.describe()
```

```python
df['Survived'].value_counts().plot(kind = 'bar')
plt.ylabel('count')
plt.title('Survival count')
plt.show()
```

```python
sns.histplot(df['Age'],kde = 'True')
plt.title('Age Distribution')
plt.show()
```

```python
for c in ["Sex", "Pclass", "Embarked"]:
    plt.figure()
    sns.countplot(x=c, data=df, order=df[c].value_counts().index)
    plt.title(f"Counts of {c}")
    plt.xticks(rotation=20)
    plt.show()
```

## Bivariate Analysis

```python
target = 'Survived'
df[target].value_counts(normalize = True)
```

```python
sns.barplot(x = 'Survived', y= 'Age',data = df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.barplot(x = 'Survived', y= 'Sex',data = df)
plt.title('Age vs Survival')
plt.show()
```

```python
sns.boxplot(x = 'Survived',y = 'Age',data = df)
plt.title('Age vs Survival boxplot')
plt.show()
```

## Multivariate Analysis

```python
sns.catplot(x="Survived", y="Age", hue="Pclass", kind="bar", data=df)
plt.title('Age vs Survived by Class')
plt.show()
```

```python
sns.pairplot(df[['Survived','Age','Fare','Pclass']])

plt.show()
```

## Outliers

```python
sns.boxplot(y = df['Age'])
plt.title('Boxplot of Age')
plt.show()
```
